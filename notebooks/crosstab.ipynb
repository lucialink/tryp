<h2> How to create a Crosstabulated DataFrame with subtotals and grand totals. </h2>

```python
import pandas as pd                                                                                                                                                                     
from tryp.crosstab import Crosstab
from IPython.display import HTML

def pprint(s):
    return HTML("<div style='overflow:auto'>%s</div>" %s)
    
df = pd.read_csv('../tests/data/fixture.csv')                                                                                                                                        
index = ['region', 'area', 'distributor']                                                                                                                                            
columns = ['salesrep', 'retailer']                                                                                                                                                   
values = ['sales', 'invoice_count']                                                                                                                                                  
index_totals = ['region', 'area']                                                                                                                                                    
columns_totals = ['region', 'area', 'distributor']

ct = Crosstab(columns, index, values, columns_totals, index_totals, df)
ct.dataframe.to_pickle('crosstab.df')

pprint(ct.dataframe.to_html())
```
